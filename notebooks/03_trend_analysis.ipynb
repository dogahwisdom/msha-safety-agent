# Step 3: Trend analysis tool

Aggregates injury counts by year, group, and period. **2026 is excluded** from year-over-year counts (partial reporting year).

**CLI/tests:** `pytest tests/test_trends.py -v`

```python
import sys
from pathlib import Path

ROOT = Path.cwd() if (Path.cwd() / "src").exists() else Path.cwd().parent
sys.path.insert(0, str(ROOT))

import matplotlib.pyplot as plt
import pandas as pd

from src.tools.trends import PARTIAL_CALENDAR_YEAR, TrendAnalyzer
```

```python
analyzer = TrendAnalyzer()

fatality_counts = analyzer.count_by_year({"degree_code": "01"})
print(f"Notes: {fatality_counts.notes}")
fatality_counts.data.tail(10)
```

```python
df = fatality_counts.data
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(df["CAL_YR"], df["injury_count"], marker="o")
ax.axvline(PARTIAL_CALENDAR_YEAR, color="red", linestyle="--", label=f"{PARTIAL_CALENDAR_YEAR} excluded from YoY")
ax.set_title("MSHA fatalities (degree code 01) by year")
ax.set_xlabel("Calendar year")
ax.set_ylabel("Count")
ax.legend()
plt.tight_layout()
plt.show()
```

```python
comparison = analyzer.compare_periods(
    (2005, 2009), (2010, 2014), filters={"degree_code": "01"}
)
comparison.data
```
